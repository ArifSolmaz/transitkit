# TransitKit Quick Start

This notebook demonstrates basic TransitKit usage.

```python
import numpy as np
import transitkit as tk
from transitkit.core import generate_transit_signal_mandel_agol, find_transits_bls_advanced, add_noise
```

```python
# Generate synthetic transit
time = np.linspace(0, 30, 2000)
flux = generate_transit_signal_mandel_agol(time, period=5.0, depth=0.01)
flux_noisy = add_noise(flux, noise_level=0.001)
```

```python
# Detect transit
result = find_transits_bls_advanced(time, flux_noisy)
print(f"Period: {result['period']:.4f} days")
print(f"SNR: {result['snr']:.1f}")
```
